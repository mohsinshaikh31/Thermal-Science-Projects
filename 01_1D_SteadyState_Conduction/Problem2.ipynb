```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# given
k = 40.0       # W/(m·C)
g0 = 1e6       # W/m^3
L = 0.1        # m
```

```python
# analytical T(x)
x = np.linspace(0, L, 201)
T = (g0/(2*k))*(L**2 - x**2)

T0 = T[0]
qL = g0 * L

print("T(0) =", T0, "°C")
print("q(L) =", qL, "W/m^2")
```

```python
plt.plot(x, T, lw=2)
plt.xlabel('x (m)')
plt.ylabel('Temperature (°C)')
plt.title('Temperature distribution (uniform volumetric generation, insulated at x=0)')
plt.grid(True)
plt.show()
```
