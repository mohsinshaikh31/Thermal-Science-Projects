```python
import numpy as np
```

```python
# Given data
k = 0.69        # Thermal conductivity [W/(m·°C)]
L = 0.20        # Thickness [m]
T1 = 30.0       # Hot surface temperature [°C]
T2 = -20.0      # Cold surface temperature [°C]
```

```python
# Calculate temperature difference
delta_T = T1 - T2
```

```python
# Steady-state heat flux: q = k * (T1 - T2) / L
q = k * delta_T / L

# Print result with formatting
print(f"Steady-state heat flux: {q:.1f} W/m²")
```
